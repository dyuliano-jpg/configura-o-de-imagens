```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

fig, ax = plt.subplots()

x = np.arange(0, 2*np.pi, 0.01)
line, = ax.plot(x, np.sin(x))

def animate(i):
    line.set_ydata(np.sin(x + i/10.0))
    return line,

ani = animation.FuncAnimation(fig, animate, np.arange(1, 200), interval=25, blit=True)

plt.show()
```

```python
import time
from IPython.display import clear_output

linhas = int(input("Tamanho da arvore? "))
tamanho = linhas * 2 - 1
espacos = (tamanho - 1) / 2

while True:
    for i in range(1, linhas+1):
        print(" " * int(espacos - i + 1), "*" * (2 * i - 1))
    print(" " * int(espacos), "|")
    time.sleep(0.5)  # pausa por meio segundo
    clear_output(wait=True)  # limpa a saída
```

```python
import time
from IPython.display import clear_output

linhas = int(input("Tamanho da arvore? "))
tamanho = linhas * 2 - 1
espacos = (tamanho - 1) / 2
pisca = False

while True:
    for i in range(1, linhas+1):
        if pisca and i % 2 == 0:
            print(" " * int(espacos - i + 1), "o" * (2 * i - 1))
        else:
            print(" " * int(espacos - i + 1), "*" * (2 * i - 1))
    print(" " * int(espacos), "|")
    time.sleep(0.5)  # pausa por meio segundo
    clear_output(wait=True)  # limpa a saída
    pisca = not pisca
```

```python
import time
from IPython.display import clear_output

linhas = int(input("Tamanho da arvore? "))
tamanho = linhas * 2 - 1
espacos = (tamanho - 1) // 2  # Use a divisão inteira
pisca = False

while True:
    clear_output(wait=True)  # Limpe a saída no início de cada loop
    if pisca:
        print(" " * int(espacos), "Feliz Natal")
    for i in range(1, linhas+1):
        if pisca and i % 2 == 0:
            print(" " * int(espacos - i + 1), "o" * (2 * i - 1), end="")
        else:
            print(" " * int(espacos - i + 1), "*" * (2 * i - 1), end="")
        if i == linhas // 2:  # Adicione a mensagem no meio da árvore
            print("   FELIZ NATAL!!! E PRÓSPERO 2024")
        else:
            print()
    print(" " * int(espacos), "|")
    time.sleep(0.5)  # Pausa por meio segundo
    pisca = not pisca  # Alterne o piscar
```
